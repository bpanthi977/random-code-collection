```python
import gym
import numpy as np

from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv, SubprocVecEnv
from stable_baselines3.common.env_util import make_vec_env
```

```python
env_id = "CartPole-v1"
num_cpu = 4  
# Create the vectorized environment
env = make_vec_env(env_id, num_cpu, vec_env_cls=SubprocVecEnv)
```

```python
model = PPO("MlpPolicy", env, verbose=1,device='cpu')
model.learn(total_timesteps=25_0000)
```

```python
def render(model, steps = 1000):
    env_single = make_vec_env(env_id, 1, vec_env_cls=DummyVecEnv)
    obs = env_single.reset()
    for _ in range(1000):
        action, _states = model.predict(obs)
        obs, rewards, dones, info = env_single.step(action)
        env_single.render()
```

```python
render(model, 500)
```
